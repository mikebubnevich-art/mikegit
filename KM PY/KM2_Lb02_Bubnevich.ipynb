```python
#нарезание и конкатенация
#создаем новую строку, взяв часть старой и соединив с новым символом
S = 'spam'
S = S[:1] + 'l' + S[2:]
print(S)
```

 **Список** - это последовательность, которую можно изменять после ее создания. Поэтому в отличие от строки, первый и второй подход будут работать. 

```python
L = ['s', 'p', 'a', 'm']
L = L[:1] + ['l'] + L[2:]
print(L)
```

```python
L = ['s', 'p', 'a', 'm']
L[1] = 'l'
print(L)
```

```python
L = L[0] + 'l' + L[2] + L[3]
print(L)
```

При индексации и конкатенации получается строка, поэтому корректнее будет просто создание нового списка.

**Кортеж** относится к неизменяемым типам, и попытка присвоить значение его элементу вызывает <span style = "color:red;"> TypeError: 'tuple' object does not support item assignment</span>

```python
T = ('s', 'p', 'a', 'm')
T = T[:1] + ('l',) + T[2:]
print(T) 
```

```python
T = (T[0],) + ('l',) + T[2:]
print(T) 
```

```python
#Соединение списка строк в одну строку с использованием разделителя
words = ['s', 't', 'r', 'i', 'n', 'g']
separator = '+'
joined_string = separator.join(words)
print(joined_string)
```

**Спецификация join**

**Join** - объединяет элементы переданной последовательности (строки, символы) в одну новую строку, вставляя между ними указанный разделитель.

Как и все строковые методы, не изменяет исходный объект, а возвращает новый строковый объект. 

**Синтаксис**: разделитель.join(последовательность)

```python
#Разбиение строки на список подстрок по разделителю
#(метод split, который в качестве аргумента принимает разделитель)
string = 's+t+r+i+n+g'
separator = '+'
splitted_list = string.split(separator)
print(splitted_list)
```

**Спецификация split**

**Split** - pазбивает исходную строку на подстроки по указанному разделителю.
Возвращает новый список подстрок.

**Синтаксис**: строка.split(разделитель)

Аргумент key позволяет задать функцию, которая будет применена к каждому элементу списка перед сравнением для сортировки. Сортировка будет выполняться на основе значений, возвращаемых этой функцией.

```python
import copy
original = [1,[2,3],4]
sh_copy = original.copy() #поверхностная копия
deep_copy = copy.deepcopy(original) #глубокая копия

original[1][0] = 'X'
print("Оригинал:", original)       
print("sh_copy:", sh_copy) 
print("deep_copy:", deep_copy)      
```

**copy()** создает овый объект, но вкладывает в него ссылки на элементы исходного объекта.

**deepcopy()** создает новый объект и рекурсивно создает копии всех вложенных объектов. Изменения в оригинале не влияют на глубокую копию.

**Основные правила использования пробелов**


1. Вокруг бинарных операторов (как =, ==, <, +, -, *, /, in, is, and, or):
- Ставить по одному пробелу с каждой стороны: x = y + 5, if x > 5 and x < 10:
- Исключение: для указания приоритета можно не ставить пробелы: x = x*2 - 1

2. После запятых, двоеточий и точек с запятой:
- Ставьте пробел после: def func(a, b, c):, my_list = [1, 2, 3]
- Не ставьте пробел перед ними: func(a , b , c) — неправильно

3. В срезах (slice):
- Одинаково хорошо: ham[1:9], ham[1:9:3], ham[:9:3]
- Пробелы допустимы, если они помогают: ham[1:9 : 3] (для сложных выражений)

**5.Назовите четыре операции, которые изменяют списковый объект на месте.**

Методы append и extend увеличивают список на месте, методы sort и reverse
упорядочивают и обращают списки, метод insert вставляет элемент по смещению, методы remove и pop удаляют элемент из списка по значению и по позиции, оператор del удаляет элемент или срез, а операторы присваивания по индексам и по срезам заменяют элемент или целую секцию. Для ответа на данный
вопрос выберите любые четыре операции из перечисленных.


```python
#append, extend, sort, reverse
lst = [1, 2, 3]
lst.append(4) #append(x) добавляет x как один элемент
print(lst)
```

```python
lst = [1, 2, 3]
lst.extend([4, 5, 6]) #extend(seq) добавляет каждый элемент из seq
print(lst)
```

```python
lst = [3, 1, 4, 1, 5]
lst.sort() #sort() сортирует список
print(lst)
```

```python
lst = [1, 2, 3, 4, 5]
lst.reverse() #reverse() переворачивает список
print(lst)
```

**6.Что принимается по умолчанию для аргумента режима обработки в вызове
open?**

По умолчанию для аргумента режима обработки в вызове open принимается ’ г ’, т.е. чтение текстового ввода. Для входных текстовых файлов нужно
просто передать имя внешнего файла.

```python
#'r' (чтение, текст)
with open('tmp.txt', 'w') as f:
    f.write("Hello, world!\nIt's a nice day today!")
```

```python
file = open('tmp.txt', 'r')
print(file.read())
file.close()
```
